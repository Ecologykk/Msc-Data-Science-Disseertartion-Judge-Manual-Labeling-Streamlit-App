# Reducing temporarily data for the app

Because judges are skeptic on the amount of time it will take. They are aiming for 5 each branch.

```python
from pathlib import Path
import sys
import pandas as pd
import numpy as np


APP_DIR = Path().resolve().parent
BOC_DIR = APP_DIR / "data" / "boc_gold_test_full.csv"
DV_DIR = APP_DIR / "data" / "dv_gold_test_full.csv"
SEED = 25
```

```python
boc_df = pd.read_csv(BOC_DIR)
dv_df = pd.read_csv(DV_DIR)
```

```python
# sampling 5
boc_sample = boc_df.sample(n=5, random_state=SEED)
dv_sample = dv_df.sample(n=5, random_state=SEED)
```

```python
BOC_SAMPLE = APP_DIR / "data" / "boc_5_sample.csv"
DV_SAMPLE = APP_DIR / "data" / "dv_5_sample.csv"

boc_sample.to_csv(BOC_SAMPLE)
dv_sample.to_csv(DV_SAMPLE)
```
